```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_vk = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=1)
df_mm = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=2)
df_ok = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=3)
df_fb_ru = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=4)
df_fb_eng = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=5)
df_fs = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=6)
df_total = pd.read_excel(r"\\megafon.ru\fr\MF\CNT\VOL03\dmitry.Kalinin\Desktop\задание 1.xlsx", sheet_name=7)
```

```python
dfs = []
dfs.append(df_vk)
dfs.append(df_mm)
dfs.append(df_ok)
dfs.append(df_fb_ru)
dfs.append(df_fb_eng)
dfs.append(df_fs)
dfs.append(df_total)
```

```python
df_names = ['VK', 'MM', 'OK', 'FB_RU', 'FB_EN', 'FS', 'TOTAL']
# Создаем график
plt.figure(figsize=(12, 6))
# Добавляем линии для каждого датафрейма
for i, df in enumerate(dfs):
    df = df.reset_index() # Сбрасываем индекс
    sns.lineplot(data=df, x='index', y='ctr_new', label=df_names[i])
# Настраиваем график
plt.title('Значения ctr_new для 7 датафреймов')
plt.xlabel('Дни')
plt.ylabel('ctr_new')
plt.legend(title='Датафреймы')
plt.show()

df_names = ['VK', 'MM', 'OK', 'FB_RU', 'FB_EN', 'FS', 'TOTAL']
# Создаем график
plt.figure(figsize=(12, 6))
# Добавляем линии для каждого датафрейма
for i, df in enumerate(dfs):
    df = df.reset_index() # Сбрасываем индекс
    sns.lineplot(data=df, x='index', y='Sent', label=df_names[i])
# Настраиваем график
plt.title('Значения Sent для 7 датафреймов')
plt.xlabel('Дни')
plt.ylabel('ctr_new')
plt.legend(title='Датафреймы')
plt.show()

df_names = ['VK', 'MM', 'OK', 'FB_RU', 'FB_EN', 'FS', 'TOTAL']
# Создаем график
plt.figure(figsize=(12, 6))
# Добавляем линии для каждого датафрейма
for i, df in enumerate(dfs):
    df = df.reset_index() # Сбрасываем индекс
    sns.lineplot(data=df, x='index', y='Hits', label=df_names[i])
# Настраиваем график
plt.title('Значения Hits для 7 датафреймов')
plt.xlabel('Дни')
plt.ylabel('ctr_new')
plt.legend(title='Датафреймы')
plt.show()
```

Данные представляют собой статистику отправленных (Sent), доставленных (Hits) push-уведомлений и CTR  по различным социальным сетям (VK, MM, OK, FB_RU, FB_EN, FS).
